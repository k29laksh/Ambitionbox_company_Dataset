```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
final=pd.DataFrame()

for j in range(1,11):
  url='https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get(url,headers=headers).text
  soup=BeautifulSoup(webpage, 'lxml')

  company= soup.find_all('div',itemscope="itemscope")


  name=[i.find('h2').text.strip() for i in company]
  rating = [i.find('span', class_="companyCardWrapper__companyRatingValue").text.strip() for i in company]
  reviews = [i.find_all('span', class_="companyCardWrapper__ActionCount")[0].text.strip() for i in company]
  salary = [i.find_all('span', class_="companyCardWrapper__ActionCount")[1].text.strip() for i in company]
  try:
    jobs = [i.find_all('span', class_="companyCardWrapper__ActionCount")[3].text.strip() for i in company]
  except:
    jobs.append(np.nan)




  dict={'name':name,'rating':rating,'reviews':reviews,'salary':salary,'jobs':jobs}
  df=pd.DataFrame(dict)

  final=final.append(df.drop_duplicates(),ignore_index=True)




```
